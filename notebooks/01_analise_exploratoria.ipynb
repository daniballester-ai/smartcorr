# Análise Exploratória SmartCorr (Arquitetura MLOps)

Este notebook analisa os dados gerados pelo pipeline MLOps.
Certifique-se de ter executado `executar_modular.bat` antes para gerar os arquivos CSV.

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Caminhos definidos no params.yaml
RAW_PATH = '../data/raw/raw_history.csv'
PROCESSED_PATH = '../data/processed/train_data.csv'
PREDICTIONS_PATH = '../data/processed/predictions.csv'
```

## 1. Análise dos Dados Brutos

```python
if os.path.exists(RAW_PATH):
    df_raw = pd.read_csv(RAW_PATH)
    print(f"Dados Brutos: {df_raw.shape}")
    display(df_raw.head())
else:
    print("Arquivo bruto não encontrado. Execute o pipeline primeiro.")
```

## 2. Análise de Features (Processados)

```python
if os.path.exists(PROCESSED_PATH):
    df_proc = pd.read_csv(PROCESSED_PATH)
    print(f"Dados Processados: {df_proc.shape}")
    display(df_proc.head())
    
    # Correlação
    plt.figure(figsize=(10, 8))
    plt.matshow(df_proc.corr(), fignum=1)
    plt.xticks(range(df_proc.select_dtypes(['number']).shape[1]), df_proc.select_dtypes(['number']).columns, fontsize=10, rotation=45)
    plt.yticks(range(df_proc.select_dtypes(['number']).shape[1]), df_proc.select_dtypes(['number']).columns, fontsize=10)
    plt.colorbar()
    plt.title('Matriz de Correlação', fontsize=12)
    plt.show()
else:
    print("Dados processados não encontrados.")
```

## 3. Análise das Previsões

```python
if os.path.exists(PREDICTIONS_PATH):
    df_pred = pd.read_csv(PREDICTIONS_PATH)
    print(f"Previsões: {df_pred.shape}")
    
    # Histograma do NS Previsto
    plt.figure(figsize=(10, 6))
    df_pred['Nivel_Servico_Predito'].hist(bins=30)
    plt.title('Distribuição do Nível de Serviço Previsto')
    plt.xlabel('Nível de Serviço')
    plt.show()
else:
    print("Previsões não encontradas.")
```
